```python
import os
import shutil
from sklearn.model_selection import train_test_split
from tqdm import tqdm

# Define paths to original training data and the new directories
original_train_dir = 'data/FERPlus/train/'
train_split_dir = 'data/FERPlusNew/'

# Create directories for training and validation
os.makedirs(train_split_dir, exist_ok=True)
os.makedirs(os.path.join(train_split_dir, 'train'), exist_ok=True)
os.makedirs(os.path.join(train_split_dir, 'val'), exist_ok=True)

# List all the subdirectories (categories) in the original train directory
categories = os.listdir(original_train_dir)

# For each category, we will split the images into training and validation sets
for category in tqdm(categories):
    # Define the original and destination paths
    category_path = os.path.join(original_train_dir, category)
    train_category_dir = os.path.join(train_split_dir, 'train', category)
    val_category_dir = os.path.join(train_split_dir, 'val', category)
    
    # Create category subdirectories in train/val directories
    os.makedirs(train_category_dir, exist_ok=True)
    os.makedirs(val_category_dir, exist_ok=True)
    
    # Get all images in the category
    image_files = os.listdir(category_path)
    
    # Split the images into train and val
    train_files, val_files = train_test_split(image_files, test_size=0.1, random_state=42)

    # Move the files into the new directories
    for file in train_files:
        shutil.copy(os.path.join(category_path, file), os.path.join(train_category_dir, file))
    
    for file in val_files:
        shutil.copy(os.path.join(category_path, file), os.path.join(val_category_dir, file))

print("Training and validation split completed.")
```
