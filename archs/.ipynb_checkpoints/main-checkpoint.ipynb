```python
train_path = BASE_PATH + "data/FER13New/train"
val_path = BASE_PATH + "data/FER13New/val"
test_path = BASE_PATH + "data/FER13New/test"

train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=30,
    width_shift_range=0.05,
    height_shift_range=0.05,
    shear_range=0.05,
    zoom_range=0.05,
    horizontal_flip=True,
    fill_mode='nearest')

val_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)

train_generator = train_datagen.flow_from_directory(train_path,
    target_size=(64, 64),
    color_mode='rgb',
    batch_size=256,
    class_mode='categorical',
    shuffle=True)

val_generator = val_datagen.flow_from_directory(val_path,
    target_size=(64, 64),
    color_mode='rgb',
    batch_size=256,
    class_mode='categorical',
    shuffle=False)

test_generator = test_datagen.flow_from_directory(test_path,
    target_size=(64, 64),
    color_mode='rgb',
    batch_size=256,
    class_mode='categorical',
    shuffle=False) # 'spare' for category
```

```python
def plot_training_history(history):
    plt.figure(figsize=(12, 5))

    # Plot accuracy
    plt.subplot(1, 2, 1)
    plt.plot(history.history['accuracy'], label='Train Accuracy')
    plt.plot(history.history['val_accuracy'], label='Val Accuracy')
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.legend()
    plt.title('Training and Validation Accuracy')

    # Plot loss
    plt.subplot(1, 2, 2)
    plt.plot(history.history['loss'], label='Train Loss')
    plt.plot(history.history['val_loss'], label='Val Loss')
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.legend()
    plt.title('Training and Validation Loss')

    plt.show()
```

```python
class MLP(layers.Layer):
    def __init__(self, in_features, hidden_features=None, out_features=None, drop_rate=0.0):
        super().__init__()
        out_features = out_features or in_features
        hidden_features = hidden_features or in_features
        self.fc1 = layers.Dense(hidden_features, activation=tf.nn.gelu)
        self.drop1 = layers.Dropout(drop_rate)
        self.fc2 = layers.Dense(out_features)
        self.drop2 = layers.Dropout(drop_rate)

    def call(self, x):
        x = self.fc1(x)
        x = self.drop1(x)
        x = self.fc2(x)
        x = self.drop2(x)
        return x

class CrossAttention(layers.Layer):
    def __init__(self, embed_dim=768):
        super().__init__()
        self.norm1 = layers.LayerNormalization()
        self.norm2 = layers.LayerNormalization()
        self.proj = layers.Dense(embed_dim * 2)
        self.mlp = MLP(embed_dim, hidden_features=int(embed_dim * 4))
        self.theta = self.add_weight(shape=(), initializer="zeros", trainable=True)
    
    def call(self, x_a, x_p, x_n, x_n2):
        x_a, x_p, x_n, x_n2 = self.norm1(x_a), self.norm1(x_p), self.norm1(x_n), self.norm1(x_n2)
        
        def project_reshape(x):
            qkv = self.proj(x)
            qkv = tf.reshape(qkv, (tf.shape(qkv)[0], tf.shape(qkv)[1], 2, -1))
            return tf.transpose(qkv, perm=[2, 0, 1, 3])
        
        QK1, V1 = project_reshape(x_a)
        QK2, V2 = project_reshape(x_p)
        QK3, V3 = project_reshape(x_n)
        QK4, V4 = project_reshape(x_n2)
        
        k = tf.tanh(self.theta)
        cross_map1, cross_map2, cross_map3, cross_map4 = self.compute_attention(QK1, QK2, QK3, QK4, k)
        
        def apply_attention(x, V, cross_map):
            cross_map = tf.reshape(cross_map, (tf.shape(x)[0], tf.shape(x)[1], 1))
            return x + cross_map * V
        
        x_a = apply_attention(x_a, V1, cross_map1)
        x_p = apply_attention(x_p, V2, cross_map2)
        x_n = apply_attention(x_n, V3, cross_map3)
        x_n2 = apply_attention(x_n2, V4, cross_map4)
        
        x_a = self.norm2(x_a) + self.mlp(x_a)
        x_p = self.norm2(x_p) + self.mlp(x_p)
        x_n = self.norm2(x_n) + self.mlp(x_n)
        x_n2 = self.norm2(x_n2) + self.mlp(x_n2)
        
        return x_a, x_p, x_n, x_n2, k
    
    def compute_attention(self, QK1, QK2, QK3, QK4, k):
        # Placeholder function: Replace with real attention computation
        return QK1, QK2, QK3, QK4

# The main model class would include components like Vision Transformer and backbone, similar to PyTorch implementation.

class PyramidTransExpr(keras.Model):
    def __init__(self, img_size=48, num_classes=8, embed_dim=768):
        super().__init__()
        self.vision_transformer = layers.Dense(embed_dim)  # Placeholder for actual ViT implementation
        self.cross_attention = CrossAttention(embed_dim)
        self.classifier = layers.Dense(num_classes, activation='softmax')

    def call(self, x_a, x_p, x_n, x_n2):
        x_a = self.vision_transformer(x_a)
        x_p = self.vision_transformer(x_p)
        x_n = self.vision_transformer(x_n)
        x_n2 = self.vision_transformer(x_n2)
        
        x_a, x_p, x_n, x_n2, k = self.cross_attention(x_a, x_p, x_n, x_n2)
        
        out_a = self.classifier(x_a)
        out_p = self.classifier(x_p)
        out_n = self.classifier(x_n)
        out_n2 = self.classifier(x_n2)
        
        return out_a, out_p, out_n, out_n2, k
```

```python
class SEBlock(keras.layers.Layer):
    def __init__(self, in_channels, reduction=16):
        super(SEBlock, self).__init__()
        self.global_avg_pool = layers.GlobalAveragePooling2D()
        self.fc1 = layers.Dense(in_channels // reduction, activation='relu', use_bias=False)
        self.fc2 = layers.Dense(in_channels, activation='sigmoid', use_bias=False)
    
    def call(self, inputs):
        x = self.global_avg_pool(inputs)
        x = self.fc1(x)
        x = self.fc2(x)
        x = tf.reshape(x, (-1, 1, 1, x.shape[-1]))
        return inputs * x

class BaseSE(keras.Model):
    def __init__(self):
        super(BaseSE, self).__init__()
        self.conv1 = layers.Conv2D(64, kernel_size=3, padding='same', activation=None)
        self.bn1 = layers.BatchNormalization()
        self.conv2 = layers.Conv2D(128, kernel_size=3, padding='same', activation=None)
        self.bn2 = layers.BatchNormalization()
        self.conv3 = layers.Conv2D(256, kernel_size=3, padding='same', activation=None)
        self.bn3 = layers.BatchNormalization()
        self.conv4 = layers.Conv2D(256, kernel_size=3, padding='same', activation=None)
        self.bn4 = layers.BatchNormalization()
        self.conv5 = layers.Conv2D(256, kernel_size=3, padding='same', activation=None)
        self.bn5 = layers.BatchNormalization()
        self.se1 = SEBlock(256)
        
        self.pool = layers.GlobalAveragePooling2D()
        self.fc1 = layers.Dense(512, activation='relu')
        self.fc2 = layers.Dense(1024, activation='relu')
        self.fc3 = layers.Dense(256, activation='relu')
        self.dropout1 = layers.Dropout(0.2)
        self.dropout2 = layers.Dropout(0.5)
        self.fc4 = layers.Dense(8, activation='softmax')
    
    def call(self, inputs):
        x = self.conv1(inputs)
        x = self.bn1(x)
        x = tf.nn.relu(x)
        x = layers.MaxPooling2D(pool_size=2)(x)
        x = self.dropout1(x)

        x = self.conv2(x)
        x = self.bn2(x)
        x = tf.nn.relu(x)
        x = layers.MaxPooling2D(pool_size=2)(x)
        x = self.dropout1(x)

        x = self.conv3(x)
        x = self.bn3(x)
        x = tf.nn.relu(x)
        x = layers.MaxPooling2D(pool_size=2)(x)
        x = self.dropout1(x)

        x = self.conv4(x)
        x = self.bn4(x)
        x = tf.nn.relu(x)
        x = layers.MaxPooling2D(pool_size=2)(x)
        x = self.dropout1(x)

        x = self.conv5(x)
        x = self.bn5(x)
        x = tf.nn.relu(x)
        x = layers.MaxPooling2D(pool_size=2)(x)
        
        x = self.se1(x)
        x = self.pool(x)
        
        x = self.fc1(x)
        x = self.dropout2(x)
        x = self.fc2(x)
        x = self.dropout2(x)
        x = self.fc3(x)
        x = self.dropout2(x)
        x = self.fc4(x)
        
        return x
```

```python
import time
from tensorflow.keras.callbacks import Callback

class CustomCallback(Callback):
    def __init__(self):
        super().__init__()
        self.epoch_start_time = None  # Initialize the variable for the start time

    def on_epoch_begin(self, epoch, logs=None):
        # Track the start time at the beginning of each epoch
        self.epoch_start_time = time.time()

    def on_epoch_end(self, epoch, logs=None):
        # Ensure that 'logs' is not None
        if logs is None:
            logs = {}

        # Calculate the time taken for the epoch
        epoch_time = time.time() - self.epoch_start_time

        # Print metrics for the first epoch and every 20th epoch
        if epoch == 0 or (epoch + 1) % 20 == 0:
            print(f"Epoch {epoch+1}: "
                  f"loss={logs.get('loss', 'N/A'):.4f}, accuracy={logs.get('accuracy', 'N/A'):.4f}, "
                  f"val_loss={logs.get('val_loss', 'N/A'):.4f}, val_accuracy={logs.get('val_accuracy', 'N/A'):.4f}, "
                  f"time={epoch_time:.2f}s")
```

```python
def train_FER(model, learning_rate, epoch, batch_size, verbose=False):
    tf.keras.backend.clear_session()
    #model.load_weights('weights\FER13New\ResEmoteNet\initial_10.h5', skip_mismatch=False)
    optimizer = Adam(learning_rate=learning_rate) #SGD(learning_rate=learning_rate, momentum=0.9)#, clipvalue=1.0) #Adam(learning_rate=learning_rate)
    model.compile(optimizer=optimizer, loss='categorical_crossentropy', metrics=['accuracy'])
    model_history = model.fit(train_generator, validation_data=val_generator, epochs=epoch, batch_size=batch_size,
                            verbose=verbose, steps_per_epoch=train_generator.samples//train_generator.batch_size,
                            validation_steps=val_generator.samples // val_generator.batch_size, callbacks=[CustomCallback()])#, callbacks=callbacks(model_name, i, Y_name, mode=mode))
    
    loss, accuracy = model.evaluate(test_generator, verbose=0)

    history_df = pd.DataFrame(model_history.history)
    history_df['test_loss'] = loss
    history_df['test_accuracy'] = accuracy
    
    #model.save_weights('weights\FER13New\ResEmoteNet\initial_11.h5', overwrite=True)
    plot_training_history(model_history)
```

```python
model = ResEmoteNet.ResEmoteNet(num_class=7)        # Val:accuracy: 0.7381     Test accuracy: 0.7381    (SGD, 3e-2, 80*7, 32, Image Generator, One Hot Encoded)
#model = CustomCNN()                     # Val: 0.7978, Test: 0.7946   (3e-4, 100, 64)
#model = PyramidTransExpr()              # Val:
#model = BaseSE()


input_tensor = tf.keras.Input(shape=(64, 64, 3))  # Adjust shape if needed
output_tensor = model(input_tensor)
model = tf.keras.Model(inputs=input_tensor, outputs=output_tensor)
model.summary()
print()

train_FER(model, 4e-3, 80*5, 256)
```
